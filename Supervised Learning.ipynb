# k-Nearest Neighbors
### k-Neighbors Classifier

```python
# importing dataset and splitting it

from sklearn.model_selection import train_test_split

import mglearn
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
print(X.shape)
print(y.shape)
```

```python
# plotting the dataset

from matplotlib import pyplot as plt

mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.title('make_forge dataset')
```

```python
# making the model and fitting the data into it

from sklearn.neighbors import KNeighborsClassifier

knnclf = KNeighborsClassifier(n_neighbors = 3).fit(X_train, y_train)
```

```python
# plotting the model

mglearn.plots.plot_2d_separator(knnclf, X, fill=True, eps=0.5, alpha=.4)
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.title('make_forge dataset')
```

```python
# testing the accuracy of the model on training set and testing set

print('Training Set Accuracy: {:.2f}'.format(knnclf.score(X_train, y_train)))
print('Testing Set Accuracy: {:.2f}'.format(knnclf.score(X_test, y_test)))
```

### Ridge Regression

```python
# importing dataset and splitting it

import mglearn
X, y = mglearn.datasets.load_extended_boston()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# creating the model and fitting in the data

from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train) # by default alpha=1.0
```

```python
# accuracy on training and test data

print('Training set score: {}'.format(ridge.score(X_train, y_train)))
print('Testing set score: {}'.format(ridge.score(X_test, y_test)))
```

```python
# testing different values of alpha

for alpha in [0.1, 1.0, 10.0]:
    model = Ridge(alpha=alpha).fit(X_train, y_train)
    print('\nAlpha = {}'.format(alpha))
    print('Training set score: {:.2f}'.format(model.score(X_train, y_train)))
    print('Testing set score: {:.2f}'.format(model.score(X_test, y_test)))
```

### Lasso Regression

```python
# importing dataset and splitting it

import mglearn
X, y = mglearn.datasets.load_extended_boston()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# creating the model and fitting data into it

from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train) # by default alpha=1.0
```

```python
# accuracy on training and test data and checking no of features considered by the model

import numpy as np
print('Training set score: {}'.format(lasso.score(X_train, y_train)))
print('Testing set score: {}'.format(lasso.score(X_test, y_test)))
print('No of feature: {} out of {}'.format(np.sum(lasso.coef_ != 0), len(lasso.coef_)))
```

```python
# testing different values of alpha

for alpha in [0.1, 1.0, 10.0]:
    model = Lasso(alpha=alpha).fit(X_train, y_train)
    print('\nAlpha = {}'.format(alpha))
    print('Training set score: {:.2f}'.format(model.score(X_train, y_train)))
    print('Testing set score: {:.2f}'.format(model.score(X_test, y_test)))
```

### Linear Models for Classification

```python
# Visualization of Decision boundary using LogisticRegression and LinearSVC

import mglearn
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=0.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title(clf.__class__.__name__)
    ax.set_xlabel('Feature 0')
    ax.set_ylabel('Feature 1')
```

```python
# Analyzing LogisticRegression

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)

logreg = LogisticRegression(C=0.01).fit(X_train, y_train)

print('Training Set Score: {:.3f}'.format(logreg.score(X_train, y_train)))
print('Testing Set SCore: {:.3f}'.format(logreg.score(X_test, y_test)))
```

### Linear Models for Multiclass Classification

```python
# loading the dataset and visualizing it

import mglearn
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.title('make_blobs dataset')
plt.legend(['Class 0', 'Class 1', 'Class 2'])
```

```python
# creating model and checking its coefficient and intercept

from sklearn.svm import LinearSVC
linsvc = LinearSVC().fit(X, y)

print('Coefficients:\n{}'.format(linsvc.coef_))
print('Intercepts:\n{}'.format(linsvc.intercept_))
```

```python
# visualizing the trained model

mglearn.plots.plot_2d_classification(linsvc, X, fill=True, alpha=0.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

import numpy as np
line = np.linspace(-15, 15)

for coef, intercept, color in zip(linsvc.coef_, linsvc.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
    
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.title('LinearSVC model visualization')
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line 0', 'Line 1', 'Line 2'], loc=(1.01, 0.3))
```

### Decision Tree Regressor

```python
import mglearn
import os
import pandas as pd

ram_prices = pd.read_csv(os.path.join(mglearn.datasets.DATA_PATH, 'ram_price.csv'))
```

```python
# visualzing the trend of the dataset

import matplotlib.pyplot as plt

plt.semilogy(ram_prices.date, ram_prices.price)
plt.xlabel('Year')
plt.ylabel('Priec in $/Mbyte')
```

```python
# splitting the data and applying log-transform

data_train = ram_prices[ram_prices.date < 2000]
data_test = ram_prices[ram_prices.date >= 2000]

import numpy as np

X_train = data_train.date[:, np.newaxis]
y_train = np.log(data_train.price)
```

```python
# creating the model and fitting the data into it

from sklearn.tree import DecisionTreeRegressor

treereg = DecisionTreeRegressor().fit(X_train, y_train)

X_all = ram_prices.date[:, np.newaxis]
pred = treereg.predict(X_all)
price_tree = np.exp(pred) # undo the log-transform
```

```python
# visualizing the fitted model

plt.semilogy(ram_prices.date, price_tree, label='Tree prediction')
plt.semilogy(data_train.date, data_train.price, label='Training Data')
plt.semilogy(data_test.date, data_test.price, label='Testing Data')
plt.xlabel('Year')
plt.ylabel('Priec in $/Mbyte')
plt.legend()
```

# Kernelized Support Vector Machine

### Linear and Nonlinear Models

```python
from mpl_toolkits.mplot3d import Axes3D, axes3d
import matplotlib.pyplot as plt
import mglearn
import numpy as np
```

```python
# importing the dataset and seperating the dependent and independent variables

from sklearn.datasets import make_blobs
X, y = make_blobs(centers=4, random_state=8)
y = y % 2
```

```python
# plotting 2D model of this dataset

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
```

```python
# plotting 3D model of the dataset

X_new = np.hstack([X, X[:, 1:]**2])
fig = plt.figure()
ax = Axes3D(fig, elev=-152, azim=-26)
mask = y==0
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60, edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel('feature 0')
ax.set_ylabel('feature 1')
ax.set_zlabel('feature 2')
```

```python
# creating model and fittng data into it

from sklearn.svm import LinearSVC
linsvc = LinearSVC().fit(X_new, y)

coef, intersept = linsvc.coef_.ravel(), linsvc.intercept_
```

```python
# plotting the model

fig = plt.figure()
ax = Axes3D(fig, elev=-152, azim=-26)

xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intersept) / -coef[2]

ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60, edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel('feature 0')
ax.set_ylabel('feature 1')
ax.set_zlabel('feature 2')
```

### Support Vector Classifier

```python
# loading the dataset

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
# splitting the training and testing data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
```

```python
# scaling the dataset (manually)

min_on_training = X_train.min(axis=0)
range_on_training = (X_train - min_on_training).max(axis=0)
X_train_scaled = (X_train - min_on_training) / range_on_training
X_test_scaled = (X_test - min_on_training) / range_on_training
```

```python
# creating the model and fitting data into it

from sklearn.svm import SVC
svc = SVC(C = 10).fit(X_train_scaled, y_train)
```

```python
# checking the score of the model

print('Training Set Score: {:.3f}'.format(svc.score(X_train_scaled, y_train)))
print('Testing Set Score: {:.3f}'.format(svc.score(X_test_scaled, y_test)))
```

# Neural Networks (Deep Learning)
### MLPClassifier

```python
# importing dataset

from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
```

```python
# visualizing dataset

from matplotlib import pyplot as plt
import mglearn

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
```

```python
# splitting training and testing dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
# manual preprocessing feature scaling

mean_on_train = X_train.mean(axis=0)
std_on_train = X_train.std(axis=0)

X_train_scaled = (X_train - mean_on_train) / std_on_train
X_test_scaled = (X_test - mean_on_train) / std_on_train
```

```python
# creating model and fitting dataset into it

from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter=1000, alpha=1, solver='lbfgs', hidden_layer_sizes=[100,100], random_state=0)
mlp.fit(X_train_scaled, y_train)
```

```python
# checking accuracy on training and testing datasets

print('Training Set Score: {}'.format(mlp.score(X_train_scaled, y_train)))
print('Testing Set Score: {}'.format(mlp.score(X_test_scaled, y_test)))
```

```python
# mpl on breast cancer dataset

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

mean_on_train = X_train.mean(axis=0)
std_on_train = X_train.std(axis=0)
X_train_scaled = (X_train - mean_on_train) / std_on_train
X_test_scaled = (X_test - mean_on_train) / std_on_train

from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter=1000, alpha=1, random_state=0)
mlp.fit(X_train_scaled, y_train)

print('Training Set Score: {:.3f}'.format(mlp.score(X_train_scaled, y_train)))
print('Testing Set Score: {:.3f}'.format(mlp.score(X_test_scaled, y_test)))
```

# Uncertainity Estimates
### Decision Function

```python
# importing dataset and training model on it

import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split

X,y = make_circles(noise=0.25, factor=0.5, random_state=1)

y_named = np.array(['Blue', 'Red'])[y]

X_train, X_test, y_train, y_test, y_named_train, y_named_test = train_test_split(X, y, y_named, random_state=0)

gbrt = GradientBoostingClassifier(random_state=0).fit(X_train, y_named_train)
```

```python
# checking the output of decision funtion

print(gbrt.decision_function(X_test)[:5])
```

```python
# the above output is what accompany the accuracy by predicting y_test

print(y_named_test[:5])
```
