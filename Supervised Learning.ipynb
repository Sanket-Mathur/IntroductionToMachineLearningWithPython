# k-Nearest Neighbors
### k-Neighbors Classifier

```python
# importing dataset and splitting it

from sklearn.model_selection import train_test_split

import mglearn
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
print(X.shape)
print(y.shape)
```

```python
# plotting the dataset

from matplotlib import pyplot as plt

mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.title('make_forge dataset')
```

```python
# making the model and fitting the data into it

from sklearn.neighbors import KNeighborsClassifier

knnclf = KNeighborsClassifier(n_neighbors = 3).fit(X_train, y_train)
```

```python
# plotting the model

mglearn.plots.plot_2d_separator(knnclf, X, fill=True, eps=0.5, alpha=.4)
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.title('make_forge dataset')
```

```python
# testing the accuracy of the model on training set and testing set

print('Training Set Accuracy: {:.2f}'.format(knnclf.score(X_train, y_train)))
print('Testing Set Accuracy: {:.2f}'.format(knnclf.score(X_test, y_test)))
```

### k-Neighbors Regressor

```python
# importing dataset and splitting it

import mglearn
X, y = mglearn.datasets.make_wave(n_samples = 40)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
# plotting the dataset

import matplotlib.pyplot as plt
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('make_wave dataset')
```

```python
# initiating the model and fitting the dataset into it

from sklearn.neighbors import KNeighborsRegressor

knnreg = KNeighborsRegressor(n_neighbors = 3).fit(X_train, y_train)
```

```python
# plotting the model

import numpy as np
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
plt.plot(line, knnreg.predict(line))
plt.plot(X, y, 'o')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('make_wave dataset')
```

# Linear Models
### Linear Regression

```python
# importing dataset and splitting it

import mglearn
X, y = mglearn.datasets.make_wave(n_samples=60)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# visualizing the dataset

import matplotlib.pyplot as plt
plt.plot(X, y, 'o')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('make_wave dataset')
```

```python
# creating model and fitting data into it

from sklearn.linear_model import LinearRegression
lreg = LinearRegression().fit(X_train, y_train)
```

```python
# printing the coefficient and intercept

print('Coefficient: {}'.format(lreg.coef_))
print('Intercept: {}'.format(lreg.intercept_))
```

```python
# accuracy on training and test data

print('Training set score: {}'.format(lreg.score(X_train, y_train)))
print('Testing set score: {}'.format(lreg.score(X_test, y_test)))
```

```python
# plotting our model

plt.plot(X, lreg.predict(X))
plt.plot(X, y, 'o')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('make_wave dataset')
```

### Ridge Regression

```python
# importing dataset and splitting it

import mglearn
X, y = mglearn.datasets.load_extended_boston()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# creating the model and fitting in the data

from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train) # by default alpha=1.0
```

```python
# accuracy on training and test data

print('Training set score: {}'.format(ridge.score(X_train, y_train)))
print('Testing set score: {}'.format(ridge.score(X_test, y_test)))
```

```python
# testing different values of alpha

for alpha in [0.1, 1.0, 10.0]:
    model = Ridge(alpha=alpha).fit(X_train, y_train)
    print('\nAlpha = {}'.format(alpha))
    print('Training set score: {:.2f}'.format(model.score(X_train, y_train)))
    print('Testing set score: {:.2f}'.format(model.score(X_test, y_test)))
```

### Lasso Regression

```python
# importing dataset and splitting it

import mglearn
X, y = mglearn.datasets.load_extended_boston()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# creating the model and fitting data into it

from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train) # by default alpha=1.0
```

```python
# accuracy on training and test data and checking no of features considered by the model

import numpy as np
print('Training set score: {}'.format(lasso.score(X_train, y_train)))
print('Testing set score: {}'.format(lasso.score(X_test, y_test)))
print('No of feature: {} out of {}'.format(np.sum(lasso.coef_ != 0), len(lasso.coef_)))
```

```python
# testing different values of alpha

for alpha in [0.1, 1.0, 10.0]:
    model = Lasso(alpha=alpha).fit(X_train, y_train)
    print('\nAlpha = {}'.format(alpha))
    print('Training set score: {:.2f}'.format(model.score(X_train, y_train)))
    print('Testing set score: {:.2f}'.format(model.score(X_test, y_test)))
```

### Linear Models for Classification

```python
# Visualization of Decision boundary using LogisticRegression and LinearSVC

import mglearn
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=0.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title(clf.__class__.__name__)
    ax.set_xlabel('Feature 0')
    ax.set_ylabel('Feature 1')
```

```python
# Analyzing LogisticRegression

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)

logreg = LogisticRegression(C=0.01).fit(X_train, y_train)

print('Training Set Score: {:.3f}'.format(logreg.score(X_train, y_train)))
print('Testing Set SCore: {:.3f}'.format(logreg.score(X_test, y_test)))
```

### Linear Models for Multiclass Classification

```python
# loading the dataset and visualizing it

import mglearn
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.title('make_blobs dataset')
plt.legend(['Class 0', 'Class 1', 'Class 2'])
```

```python
# creating model and checking its coefficient and intercept

from sklearn.svm import LinearSVC
linsvc = LinearSVC().fit(X, y)

print('Coefficients:\n{}'.format(linsvc.coef_))
print('Intercepts:\n{}'.format(linsvc.intercept_))
```

```python
# visualizing the trained model

mglearn.plots.plot_2d_classification(linsvc, X, fill=True, alpha=0.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

import numpy as np
line = np.linspace(-15, 15)

for coef, intercept, color in zip(linsvc.coef_, linsvc.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
    
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.title('LinearSVC model visualization')
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line 0', 'Line 1', 'Line 2'], loc=(1.01, 0.3))
```
